# Introduction:

The purpose of this project is to help a greek company identify an opportunity to open a new branch of its fried chicken specialty restaurant in Tokyo Japan. It will allow the company to identify the location of its main competitor KFC and where to locate its potential new store where there will be considerable foot traffic

Fried chicken is very popular in Japan, so much so that  its a christmas tradition to dine at a KFC restaurant on christmas

Problem to Solve:

The major purpose of this project, is to suggest potential good locations to open their first store in Japan .

## Data:

We will need data about different venues in different neighborhoods around Shibuya. In order to gain that information we will use "Foursquare" locational information. Foursquare is a location data provider with information about a variety of venues and events within an area of interest. Such information includes venue names, locations, menus and even photos. As such, the foursquare location platform will be used as the sole data source since all the stated required information can be obtained through the API.

After choosing the Shibuya , we then connect to the Foursquare API to gather information about venues inside each and every neighborhood. For each neighborhood, we have chosen the radius to be 100 meter.

The data retrieved from Foursquare contained information of venues within a specified distance of the longitude and latitude of the postcodes. The information obtained per venue as follows:

1. Neighborhood
2. Neighborhood Latitude
3. Neighborhood Longitude
4. Venue
5. Name of the venue e.g. the name of a store or restaurant
6. Venue Latitude
7. Venue Longitude
8. Venue Category
